```python
import matplotlib.pyplot as plt
```

```python
t=[1,2,3,4,5,6,7,8,9]
s1=[45,23,56,89,42,-8,91,76,67]
s2=[56,89,42,-8,91,76,67,78,95]
s3=[-100,78,45,23,56,89,42,-8,91]
```

```python
plt.figure(figsize=(8,9),dpi=100)

plt.subplot(3,2,1)
plt.plot(t,s1)

plt.subplot(3,2,2)
plt.scatter(t,s2)

plt.subplot(3,2,3)
plt.bar(t,s3)

plt.subplot(3,2,4)
plt.barh(t,s1)

plt.subplot(3,2,5)
plt.plot(t,s2)
plt.plot(t,s1)

plt.subplot(3,2,6)
plt.scatter(t,s1)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,9),dpi=100)
f,ax=plt.subplots(3,2)
ax[0,0].plot(t,s1,marker="v",color="k")
ax[1,0].bar(t,s1,color="g")
ax[0,1].scatter(t,s2)
ax[1,1].plot(t,s1,marker="v",color="c")
ax[2,0].stackplot(t,s1,s2,s3,colors=["r","k","c"])
ax[2,1].plot(t,s1,marker="v",color="k")
plt.tight_layout()
plt.show()
```

```python
x=[1,2,3,4,5,6,7,8,9]
y=[45,23,56,89,42,-8,91,76,67]
z=[56,89,42,-8,91,76,67,78,95]
```

```python
fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection="3d")
ax.scatter(x,y,z,color ="g",marker="D",label="data1")
ax.scatter(x,z,y,color ="r",marker="*",label="data2")
ax.set_xlabel("xaxis")
ax.set_ylabel("yaxis")
ax.set_zlabel("zaxis")
plt.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(10,10),dpi=100)
ax = fig.add_subplot(121,projection="3d")
ax.scatter(x,y,z,color ="g",marker="D",label="data1")
ax.set_xlabel("xaxis")
ax.set_ylabel("yaxis")
ax.set_zlabel("zaxis")
plt.legend()

ay = fig.add_subplot(122,projection="3d")
ay.scatter(x,y,z,color ="g",marker="D",label="data1")
ay.scatter(x,z,y,color ="r",marker="*",label="data2")
ay.set_xlabel("xaxis")
ay.set_ylabel("yaxis")
ay.set_zlabel("zaxis")
plt.legend()
plt.show()
```
